## Illustrating Bagging by fitting various different fifth order curves

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 50)
y = 2*x + 5 + 10*np.random.rand(50)
plt.scatter(x,y)
```

```python
p = np.poly1d(np.polyfit(x, y, 5))
t = np.linspace(0, 10, 2000)
plt.plot(x, y, 'o', t, p(t), '-')
# print('Mean Sqaured Error',mean_squared_error(p(t), y))
```

```python
len(x),len(y)
```

```python
import random as rd
poly_to_fit = []
for i in range(100):
    temp = rd.sample(set([i for i in range(50)]),10)
    x_ = [x[i] for i in temp]
    y_ = [y[i] for i in temp]
    p = np.poly1d(np.polyfit(x_, y_, 5))
    t = np.linspace(0, 100, 20000)
    poly_to_fit.append(p(t))
```

```python
final = np.zeros(len(poly_to_fit[0]))
```

```python
for i in poly_to_fit:
    final+=i
```

```python
final = final/len(poly_to_fit)
```

```python
# plt.plot(x[:35], y[:35], 'o', t, final, '-')
# plt.plot(x[:35], y[:35], 'o', t, p(t), '-')
plt.plot(final)
```
