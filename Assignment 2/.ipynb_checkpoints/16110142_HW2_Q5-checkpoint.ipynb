## Uniform Random Number Generator

* We make use of the XOR Shift algorithm as given in the wikipedia page [here](https://en.wikipedia.org/wiki/Xorshift#xorshift*).
* Another excellent implementation of this algorithm is [here](https://github.com/vkumar1997/PseudoRandom-Number-Generator). 

```python
## Function Definition

## XOR Shift - The main function

def xorshift(seed,start,end):
    x = seed&0xFFFFFFFFFFFFFFFF
    x^= x>>12
    x^= x<<25
    x^= x>>27
    num = x*0x2545F4914F6CDD1D
    num = num%(end-start) + start
    return num

## Helper function to call the main function and generate random numbers
def generate_random_number(start, end, number_of_samples):
    list_of_numbers = []
    for i in range(number_of_samples):
        list_of_numbers.append(xorshift(int(str(time.time()).split('.')[1]),start,end))
    return list_of_numbers
```

```python
import time
import matplotlib.pyplot as plt

random_list = generate_random_number(1,100,1000)


#plt.hist(random_list)
```

```python
plt.xlabel('Number Obtained')
plt.ylabel('Count')
plt.title('Histogram Plot for random number generator')
plt.hist(random_list)
```
